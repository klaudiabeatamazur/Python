Also, values of both, highest and lowest mean, are different in subset and full dataframe. That basically shows that subset shall not be used to determine values like mean because it can completly change the results and the comclusions. 

```python
# 4.1 Aggregating with transform()
```

```python
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df.head(25)
```

```python
# 4.2 labeling with loc()
```

```python
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# 4.3 checking result
```

```python
df['loyalty_flag'].value_counts(dropna = False)
```

```python
# 6 Creating a spending flag for each user based on the average price across all their orders
```

```python
# 6.1 grouping by prices and user id
```

```python
df['mean_price'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# 6.2 labeling with loc()
```

```python
df.loc[df['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df.loc[df['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# 6.3 checking result
```

```python
df.head(25)
```

```python
df['spending_flag'].value_counts(dropna = False)
```

```python
# 7 determining frequent versus non-frequent customers. Creating an order frequency flag according to the median.
```

```python
# 7.1 grouping by days since prior order and user id
```

```python
df['mean_days'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# 7.2  labeling with loc()
```

```python
df.loc[df['mean_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df.loc[df['mean_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df.loc[(df['mean_days'] > 10) 
& (df['mean_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# 7.3 checking result
```

```python
df['frequency_flag'].value_counts(dropna=False)
```
