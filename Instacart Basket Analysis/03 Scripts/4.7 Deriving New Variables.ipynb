```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

# step 2 - Creating a new column for using a suitable method

The clients have changed their minds about the labels already created in “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days.

```python
# showing list of busiest and slowest
df[ 'order_day_of_week' ].value_counts(dropna = False)
```

```python
# defining busiest and slowest days
```

```python
result = []

for value in df["order_day_of_week"]:
  if value == 1 or value == 0:
    result.append("Busiest days")
  elif value == 4 or value == 3:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
df['Busiest days'] = result

```

# step 4 - Creating a new column: busiest_period_of_day and labels

When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” 

```python
# identyfing busiest hours of the day
```

```python
df[ 'order_hour_of_day' ].value_counts(dropna = False)
```

```python
# defining time periods 
```

```python
result = []

for value in df["order_hour_of_day"]:
  if value == 10 or value == 11 or value == 9 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16:
    result.append("Most orders")
  elif value == 17 or value == 8 or value == 18 or value == 19 or value == 7 or value == 20 or value == 21 or value == 22:
    result.append("Average orders")
  else:
    result.append("Fewest orders")
```

above, could also be done with the loop

```python
df['busiest_period_of_day'] = result
```
